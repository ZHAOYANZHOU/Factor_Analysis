```python
df = pd.DataFrame(data=np.random.normal(0, 1, 1000),
                  columns = ['ret'])
df['ma1'] = df['ret'].rolling(window=30).mean()
df['ma2'] = df['ret'].rolling(window=10).mean()
df
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(df.index, df['ret'])
ax.plot(df.index, df['ma1'], color='red')
ax.plot(df.index, df['ma2'], color='yellow')
ax.legend(['ma1', 'ma2', 'ret'], loc='best')
plt.show()
```
